```python
import torch
```

```python

m = torch.Tensor([1, 2, 3])
n = torch.Tensor([[4],
                 [5]])
m,n
```

```python
# 
result1 = m -n
result2 = torch.sub(m, n)

result1, result2
```

```python
# 使用原地相减的方法
result3 = m.sub_(n)
result3
# 把 m - n 的结果写回 m，但 m 只有 [1, 3] 的空间，而 m - n 需要 [2, 3]，无法原地存储，所以 PyTorch 拒绝执行
```

```python
# 创建[3, 2] [4, 2]的随机数矩阵PQ
P = torch.normal(0,0.01,(3, 2))
Q = torch.normal(0,0.01,(4, 2))
P, Q
```

```python
# 对Q进行形状变换得到的Q的转置

Q_transposed = Q.t()

Q_transposed
```

```python
# 对上述得到的P和Qt求内积
inner_product = torch.dot(P, Q_transposed)
inner_product
# 两个矩阵大小不匹配，无法进行内积计算，可能要表达的是矩阵相乘？？
```

```python
# 进行矩阵相乘
inner_product = torch.mm(P, Q_transposed)
inner_product
```

```python
# 第三问
x = torch.tensor(1.0,requires_grad=True,dtype=torch.float32)
y1 = x**2
with torch.no_grad(): # 按要求这一步计算不计算梯度
    y2 = x**3
y3 = y1 + y2

y3.backward() # 反向传播求梯度
print(f'x = {x}, y3 = {y3}, y1 = {y1}, y2 = {y2},dy3/dx = {x.grad}')
```
